# One Hot Endcode

One Hot Endcode - Mengapa perlu dilakukan One Hot Endocing? One Hot Endcoding dilakukan untuk mengubah variabel kategorikal (seperti tipe data object) menjadi vektor biner yang setiap kategorinya mewakili elemen dari vektor. Nilai 1 menunjukkan data tersebut cocok dengan elemennya dan Nilai 0 menunjukkan ketidakcocokan, dilakukannya One Hot Endcode agar memudahkan komputer membaca data dalam permodelan dan analisis data, One Hot Endcode ini juga digunakan bagi data kategori yang tidak bisa dibandingkan. Sebagai contoh adalah data gender, di dalam data ini terdapat 3 buah gender yaitu 'male' , 'female' , dan 'other'. Karena gender tidak dapat dibandingkan maka dilakukan One Hot Endcode.

```python
encode = pd.get_dummies(data, prefix=None)
encode
#Fungsi pd.get_dummies() adalah fungsi yang disediakan oleh library pandas. 
#Fungsi ini digunakan untuk melakukan One Hot Encoding pada variabel kategorikal dalam suatu data. 
#Fungsi pd.get_dummies() menghasilkan data baru dengan kolom-kolom biner yang mewakili kategori unik dalam variabel kategorikal.
#Dalam kasus ini pada tabel ‘gender’ sekarang ditambahkan unique dari gender-gender tersebut.
#Hal ini berlaku juga pada tabel ‘smoking_history’ yang ditambahkan dengan unique dari tabel ‘smoking_history’ tersebut.
```

# Normalize Data

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(encode)
data_normalisasi = pd.DataFrame(d, columns=encode.columns)
data_normalisasi
```

# Split Test

Penggunaan Split test merujuk pada pembagian dataset menjadi dua bagian yaitu dataset pelatihan (training set) dan dataset pengujian (test set). Tujuan dari pembagian dataset ini untuk melakukan pelatihan model pada dataset pelatihan (training set) dan menguji model pada dataset pengujian (test set).

```python
from sklearn.model_selection import train_test_split
X = data_normalisasi.drop(['diabetes', 'hypertension', 'heart_disease'], axis=1)
y = data_normalisasi['diabetes']

# Test = 20% dan Train = 80%
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
#Kita dapat melakukan split test menggunakan library scikit-learn.
#variabel X adalah variabel yang menyimpan dataset yang sudah dinormalisasi. 
#Dan drop adalah metode menghapus kolom yang ditentukan. 
#axis=1 adalah untuk menunjukkan yang dihapus adalah kolom bukan baris.
#variabel Y menyimpan dataset yang sudah dinormalisasi.
#Lalu menggunakan fungsi 'train_test_split' untuk membagi datset memnjadi data training dan data test. 
#Penggunaan 'test_size=0,2' menunjukan pengujian terdiri dari 20% dan 80% digunakan untuk data pelatihan.
#Yang menghasilkan variabel 'X_train','X_test','y_train', dan 'y_test' 
#yang berisi data training dan data test yang telah terpisah.
```

# Training Mode - SVM

Fungsi dari melakukan Training Mode - SVM : 
Memisahkan variabel target dari variabel independen dalam dataset.
Melakukan pra-pemrosesan data, seperti mengubah variabel kategorikal menjadi numerik dengan one-hot encoding dan melakukan normalisasi data.
Membagi dataset menjadi set latih dan set uji dengan menggunakan fungsi train_test_split dari library sklearn.
Membangun model SVM dengan menggunakan fungsi svm.SVC dari library sklearn. Pada tahap ini, parameter seperti kernel, C, dan gamma dapat diatur sesuai kebutuhan.
Melatih model SVM pada set latih menggunakan metode fit() dari library sklearn.
Melakukan prediksi pada set uji dengan menggunakan metode predict() dari library sklearn.
Evaluasi performa model dengan menggunakan berbagai metrik, seperti precision, recall, f1-score, dan accuracy, yang dihitung berdasarkan hasil prediksi model pada set uji yang sebenarnya.
Jika performa model belum memuaskan, kita dapat melakukan tuning parameter pada model SVM untuk meningkatkan performa model.

```python
from sklearn import svm
model_svm = svm.SVC()
```

```python
model = model_svm.fit(X_train, y_train)
```

# Evaluasi Model - SVM

Evaluasi Model SVM memiliki fungsi yang dapat membantu mengukur sejauh mana model dapat memprediksi dengan benar. 


```python
accuracy_test_svm = model_svm.score(X_test,y_test)
```

```python
accuracy_train_svm = model_svm.score(X_train,y_train)
#model_svm.score(X_train, y_train) adalah metode yang digunakan untuk 
#menghitung skor dari model SVM yang telah dilatih menggunakan data pelatihan (X_train) dan target pelatihan (y_train). 
#Skor yang dihasilkan adalah akurasi dari model SVM terhadap data pelatihan. 
#Di dalam konteks ini, skor  yang diperoleh dari  model_svm.score() menunjukkan 
#sejauh mana model SVM dapat memprediksi dengan benar pada data pelatihan yang digunakan untuk melatih model tersebut.
#Skor tersebut dapat bernilai antara 0 dan 1, di mana nilai 1 menunjukkan akurasi yang sempurna, 
#sedangkan nilai 0 menunjukkan model tidak memprediksi dengan benar.
```

```python
print(f"Akurasi Model (Train) : {np.round(accuracy_train_svm * 100,2)} %")
print(f"Akurasi Model (Test)  : {np.round(accuracy_test_svm * 100,2)} %")
#Hasil dari Evaluasi Model SVM ini dicetak pada perintah print(f"Akurasi Model (Train) : {np.round(accuracy_train_svm * 100,2)} %") 
#dan print(f"Akurasi Model (Test)  : {np.round(accuracy_test_svm * 100,2)} %")
#np.round(accuracy_train_svm * 100, 2) digunakan untuk membulatkan persentase 
#akurasi pelatihan dari model SVM menjadi bentuk desimal dengan dua angka di belakang koma. 
#Dalam konteks ini, persentase akurasi pelatihan dikalikan dengan 100, 
#kemudian dibulatkan ke dua angka di belakang koma menggunakan fungsi np.round(). 
#Hasul dari perintah ini dicetak menjadi sebuah string dengan format : 
#“Akurasi Model (Train/Test) : (hasil dari akurasi pelatihan) %”
```
